```python
trajT = trajs_expanded[0]
plt.figure(figsize=(30,5))
r = np.sqrt(np.square(trajT[0][:,0])+np.square(trajT[0][:,1]))
t = np.arange(0,500001*1e-5,1e-5)
plt.plot(t,trajT[0][:,0])
# plt.plot(t,trajT[0][:,1])
# plt.plot(t,r)
```

```python
radis = []
RCPx=[]
RCPy=[]
LCPx=[]
LCPy=[]
for traj in trajs_expanded:
    r = np.average(np.sqrt(np.square(traj[0][:,0])+np.square(traj[0][:,1])))
    parameters = traj[1].split("_")
    if parameters[0] =="RCP":
        RCPx.append(float(parameters[1]))
        RCPy.append(r)
    if parameters[0] =="LCP":
        LCPx.append(float(parameters[1]))
        LCPy.append(r)
    
    # radis.append([r,parameters[1],parameters[0]])
    # print(traj[1],r)
print(RCPx)
print(RCPy)
print(LCPx)
print(LCPy)
```
